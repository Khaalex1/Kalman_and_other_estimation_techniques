# Complete Kalman filter and exemple of goniometric localization with landmarks

```python
import numpy as np
from numpy import cos, sin, pi
from scipy.linalg import det, expm, inv, sqrtm
import matplotlib.pyplot as plt
from matplotlib import animation
from numpy.random import randn

# to save animated plot as gif
from celluloid import Camera
```

```python
#complete filter

def kalman_correction(xc, y, C, G_x, G_b):
    yt = y - C @ xc
    G_y = C @ G_x @ C.T + G_b
    K = G_x @ C.T @ np.linalg.inv(G_y)
    xc = xc + K @ yt
    G_x = G_x - K @ C @ G_x
    return xc, G_x

def kalman_prediction(xc, G_x, A, u, G_alpha):
    xc = A@xc + u 
    G_x = A@G_x@A.T + G_alpha
    return xc, G_x

def kalman(xc, G_x, G_b, A, u, G_alpha,  y, C):
    x_c , G_x = kalman_correction(xc, y, C, G_x, G_b)
    x_c, G_x = kalman_prediction(xc, G_x, A, u, G_alpha)
    return x_c, G_x

# confidence ellipse

def ellipse(mean, G, eta):
    """
    to plot confidence ellispe
    :param mean: mean vector of distribution
    :param G: Covariance matrix of distribution
    :param eta: distribution remains inside the ellipse with proba eta
    """
    a = np.sqrt(-2 * np.log(1 - eta))
    theta = np.linspace(0, 2 * np.pi, 100)
        
    w = np.repeat(mean, 100, axis=1)
    sq = sqrtm(G)
    W = w + a*sq@np.array([np.cos(theta), np.sin(theta)])

            
    X = W[0]
    Y = W[1]
    plt.plot(X, Y, color = "pink")

```

# Ex 7.18 : Goniometric localization



```python
# geometrical intuition allows to write Z(k+1) = A.Z(k)
# and y = C.Z

def gonio():
    #Amers = landmarks
    Amers = [np.array([[0], [25]]), np.array([[15], [30]]), np.array([[30], [15]]), np.array([[15], [20]])]
    amerx = []
    amery = []
    for i in range(4):
        amerx.append(Amers[i][0,0])
        amery.append(Amers[i][1,0])

    dt = 0.1
    # time range
    T = np.arange(0, 30, dt)
    # initialization
    z = np.array([[10, 10, 4]]).T
    G_z = np.zeros((3,3))
    uz = np.array([[0], [0], [0]])
    G_alpha_z = dt * np.diag([0.01, 0.01, 0.01])
    delta = np.pi/5
    theta = 0
    camera = Camera(plt.figure())
    for i in range(len(T)):
        # plotting the moving point (robot) and its confidence ellipse and the landmarks (blue)
        if i == 0:
            plt.scatter(z[0, 0], z[1, 0], c='green', s=50)
            ellipse(z[0:2], G_z[0:2, 0:2], 0.9)
            plt.scatter(amerx, amery, color="blue", s=100)
            camera.snap()
        else:
            theta = theta + dt*(z[2,0]*np.sin(delta/3))
            A = np.array([[1, 0, dt * np.cos(delta) * np.cos(theta)],
                          [0, 1, dt * np.cos(delta) * np.sin(theta)],
                          [0, 0, 1]])
            for amer in Amers:
                # check if robot is near a landmark (complete kalman) or not (prediction mode)
                if np.linalg.norm(z[0:2]-amer)<15:
                    angle_ai = np.arctan((amer[1,0]-z[1,0])/(amer[0,0]-z[0,0]))
                    C = np.array([[-np.sin(angle_ai), np.cos(angle_ai), 0]])
                    y = C@z
                    z, G_z = kalman_correction(z, y, C, G_z, G_b = np.array([[1]]))
                    z, G_z = kalman_prediction(z, G_z, A, uz, G_alpha_z)
                    plt.plot([z[0, 0], amer[0, 0]], [z[1, 0], amer[1, 0]], color='red')
                else :
                    z, G_z = kalman_prediction(z, G_z, A, uz, G_alpha_z)
            plt.scatter(z[0, 0], z[1, 0], c='green', s=50)
            plt.scatter(amerx, amery, color="blue", s=100)
            ellipse(z[0:2], G_z[0:2, 0:2], 0.9)
            camera.snap()
    anim = camera.animate(blit=True)
    anim.save("gonio.gif")
```

```python
# trajectory gif output displayed below

gonio()
```
