# Ex 7.7



We typically have a linear equation of the form $y = C.x$, with :

$$
  C =
  \left[ {\begin{array}{cc}
    2 & 3\\
    3 & 2\\
    1 & -1\\
  \end{array} } \right]
$$
and 

$$
  y =
  \left[ {\begin{array}{c}
    8\\
    7\\
    0\\
  \end{array} } \right]
$$

Furthermore, we have assumptions over the covariance matrix $\Gamma_{\beta}$, allowing to apply the Linear Estimator / Kalman's correction

```python
# initialize xbar as (0,0) / first guess
xbar = np.array([[0],
                 [0]])

#initialize G_x as high, because we are not confident at all that xbar actually equals (0,0)
G_x = np.array([[1000,0],
                [0, 1000]])

#given
y = np.array([[8],
              [7],
              [0]])

C = np.array([[2,3],
              [3,2],
              [1,-1]])

G_b = np.array([[1,0,0],
                [0,4,0],
                [0,0,4]])

#Kalman's correction

def correction(xbar, G_x, y, C, G_b):
    G_y = C@G_x@C.T + G_b
    K = G_x@C.T@np.linalg.inv(G_y)
    yt = y - C@xbar
    xc = xbar + K@yt
    y_hat = y - C@xc
    G_eps = G_x -K@C@G_x
    return xc, y_hat, G_eps

xc, y_hat, G_eps = correction(xbar, G_x, y, C, G_b)

print("x hat = \n", xc)
print("y hat = \n", y_hat)
print("Gamma epsilon = \n", G_eps)
```

# Ex 7.9



## 1) Kalman Correction for finding p1, p2

Note that by conserving the previous formalism, 'xbar' here may be ambiguous but refers to the estimation of (p1, p2), and NOT x(t)

```python
# 1st guess
xbar = np.array([[0], [0]])
G_x = 1000 * np.identity(2)

# convert std into meters and then square it to have variance 
G_b = np.array([[0.01, 0,0,0],[0,0.01,0,0],[0,0,0.01,0],[0,0,0,0.01]])

y = np.array([[0.38],
              [3.25],
              [4.97], 
              [-0.26]])

t = np.array([1,2,3,7])
C = np.zeros((4,2))
for i in range(4):
    C[i]=np.array([1, -np.cos(t[i])])

xc, y_hat, G_eps = correction(xbar, G_x, y, C, G_b)

print("x hat = \n", xc)
print("y hat = \n", y_hat)
print("Gamma epsilon = \n", G_eps)
```

## 2) Path drawing

```python
p1,p2 = xc
t = np.linspace(0,10,1000)
x= p1*t -p2*np.sin(t)
y = p1 - p2*np.cos(t)

plt.figure()
plt.grid()
plt.plot(x,y)
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.title('Estimated path of the mass y(x)')
```
