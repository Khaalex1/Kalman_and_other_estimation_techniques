# Ex 6.2



### 1)

We will estimate the parameter vector $P_{c}$ by using the LSE, so we have to find a formulation such as :
$$A.P_{c} = Y$$

The LSE gives :

$$P_{c} = (A.T \times A)^{-1} Y$$

```python
def ls(A,Y):
    """
    Least squares estimator. Finds X approximating AX = Y with LS estimation
    :param A: numpy array (m,n)
    :param Y: numpy array (m,p)
    :return: estimated solution numpy array (n,p)
    """
    ATA = A.T @ A
    B = A.T @ Y
    return np.linalg.solve(ATA,B)
```

We initialize the Data (matrices A and Y) such as :

```python
T = [-3, -1, 0, 2, 3, 6]
Y = np.array([17,3,1,5,11,46])
A = np.zeros((6,3))
for i in range(6):
    A[i] = np.array([T[i]**2, T[i], 1])
```

P (p1, p2, p3) is given by :

```python
P = ls(A, Y)

print("[p1, p2, p3] = ", P)
```

### 2)

Filtered measurements are given by :
$$Y_{c} = A.P_{c}$$

The residual vector is given by $Y - Y_{c}$

```python
Y_c = A@P

plt.figure()
plt.grid()
plt.plot(T, Y_c, label = "filtered measured Yc", color = 'red')
plt.scatter(T, Y, label="actual measures Y", color='blue')
plt.plot(T, Y - Y_c , label = 'residual vector Y - Yc', color = 'green')
plt.legend()
plt.xlabel('t')
plt.ylabel('Y')
plt.title('Y, Y filtered and residue in function of time')
plt.show()

print('filtered measures Yc = ', Y_c)
print('residue Y - Yc = ', Y - Y_c )
```

### 1)

```python
# matricial formalism
# Y = Omega, A = concat(U, Tr)

Y = np.array([5, 10, 8, 14, 17])
U = np.array([[4, 10, 10, 13, 15]]).reshape((5, 1))
Tr = np.array([[0, 1, 5, 5, 3]]).reshape((5, 1))
A = np.hstack((U, Tr))

# computation of lse & residue

p_c = ls(A, Y)
Y_c = A @ p_c

print("[p1, p2] = ", p_c)
print("Omega filtered = ", Y_c)

```

```python
plt.figure()
plt.grid()
plt.plot(Y_c, label = "filtered measured Yc", color = 'red')
plt.scatter(np.linspace(0, 4, 5), Y, label="actual measures Y", color='blue')
plt.plot(Y - Y_c , label = 'residual vector Y - Yc', color = 'green')
plt.legend()
plt.xlabel('measures')
plt.ylabel('Y')
plt.title('Y, Y filtered and residue')
plt.show()
```
